```python
average_metrics_by_category = df.groupby('Categories')[['Subscribers', 'Visits', 'Likes', 'Comments']].mean()
```

```python

# Calculate the average values for each metric
average_subscribers = df['Subscribers'].mean()
average_visits = df['Visits'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()

# Create a new DataFrame with above-average streamers
above_average_df = df[
    (df['Subscribers'] > average_subscribers) &
    (df['Visits'] > average_visits) &
    (df['Likes'] > average_likes) &
    (df['Comments'] > average_comments)
]

# Sort the DataFrame by the metrics
top_performers = above_average_df.sort_values(by=['Subscribers', 'Visits', 'Likes', 'Comments'], ascending=False)

# Create a combined bar chart with all four metrics
plt.figure(figsize=(12,9))
plt.bar(top_performers.index, top_performers['Subscribers'], label='Subscribers')
plt.bar(top_performers.index, top_performers['Visits'], bottom=top_performers['Subscribers'], label='Visits')
plt.bar(top_performers.index, top_performers['Likes'], bottom=[sum(x) for x in zip(top_performers['Subscribers'], top_performers['Visits'])], label='Likes')
plt.bar(top_performers.index, top_performers['Comments'], bottom=[sum(x) for x in zip(top_performers['Subscribers'], top_performers['Visits'], top_performers['Likes'])], label='Comments')
plt.ylabel('Value')
plt.title('Top Performers - All Metrics')
plt.xtricks(ro)
plt.legend()
plt.show()
```
